```python
import fitz

pdf_files = [
    "indian_constitution.pdf",
    "consumer_act.pdf",
    "indian_penal_code.pdf",
    "code_of_criminal_procedure.pdf",
    "motor_vehicle_act.pdf"
]

output_file = "combined_text.txt"

with open(output_file, "w", encoding="utf-8") as outfile:
    for pdf in pdf_files:
        doc = fitz.open(pdf)
        for page in doc:
            text = page.get_text("text")
            outfile.write(text + "\n\n")
        outfile.write("\n" + "="*80 + "\n\n")  

print("All PDFs have been combined into", output_file)


with open(output_file, "r", encoding="utf-8") as infile:
    combined_text = infile.read()

print(combined_text[:1000])
```

```python
import re

def clean_text(text):
    text = re.sub(r'\n+',"\n", text)  
    text = re.sub(r'\s+', ' ', text)   
    text = re.sub(r'[^A-Za-z0-9.,;\n\s]', '', text)  
    return text.strip()

cleaned_text = clean_text(combined_text)
print(cleaned_text[:1000]) 
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_core.documents import Document


text_splitter = RecursiveCharacterTextSplitter(
    # separator="\n",
    chunk_size=4096,
    chunk_overlap=200,
    length_function=len,
    is_separator_regex=False,

)

chunks = text_splitter.split_text(cleaned_text)

```

```python
documents = [Document(page_content=chunk) for chunk in chunks]
```

```python
documents[1]
```

```python
from sentence_transformers import SentenceTransformer

# 1. Load a pretrained Sentence Transformer model
model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
import faiss
from langchain_community.docstore.in_memory import InMemoryDocstore
from langchain_community.vectorstores import FAISS


example_embedding = model.encode(chunks[0])
dimension = example_embedding.shape[0] # Get the dimension from the example.
index = faiss.IndexFlatL2(dimension)

vector_store = FAISS(model.encode, index, docstore=InMemoryDocstore(), index_to_docstore_id={})

vector_store.add_documents(documents)
```

```python
vector_store.save_local("VectoreStore")
```

```python
query = "what are the laws for child labor"

# Perform similarity search
results = vector_store.similarity_search(query, k=4)  # Retrieve top 2 results

```

```python
for res in results:
  print(res)
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
model_id = "mistralai/Mistral-7B-Instruct-v0.1"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    device_map="auto",  
    torch_dtype="auto"
)
generator = pipeline("text-generation", model=model, tokenizer=tokenizer)

```

```python
query = "what are the laws for child labor"
retrieved_context = "\n\n".join([doc.page_content for doc in results]) 


prompt = f"<s>[INST] You are a legal expert. Use the following legal context to answer the user's question.\n\nContext:\n{retrieved_context}\n\nQuestion: {query}\n\nAnswer: [/INST]"

```

```python

output = generator(
    prompt,
    max_new_tokens=512,
    do_sample=True,
    temperature=0.7,
    top_p=0.9
)

print(output[0]["generated_text"])
```
